```python
import numpy as np
import math
from PIL import Image
from numpy import asarray
import matplotlib.pyplot as plt
```

```python
def i_list(matrix):
    zero = matrix[1][2]
    one = matrix[0][2]
    two = matrix[0][1]
    three = matrix[0][0]
    four = matrix[1][0]
    five = matrix[2][0]
    six = matrix[2][1]
    seven = matrix[2][2]
    
    output_list = [zero, one, two, three, four, five, six, seven]
            
    return output_list

def delta_cal(theta):
    if theta>90:
        delta_theta = theta-180
    elif -90<=theta<=90:
        delta_theta = theta
    elif theta<-90:
        delta_theta = theta+180
        
    return delta_theta

def poincare_index(mat_list):
    delta_theta_sum = 0
    for i in range(len(mat_list)):
        #mat_list[i] = math.radians(mat_list[i])
        theta = (mat_list[(i+1)%8])-mat_list[i]
        delta_theta = delta_cal(theta)
        delta_theta_sum += delta_theta
    
    poincare = delta_theta_sum/180
    
    return poincare

def type_finder(poincare):
    if poincare==0:
        print("Not a singular point (Non-singularity)")
    elif poincare==1:
        print("Loop")
    elif poincare==-1:
        print("Delta")
    elif poincare==2:
        print("Whorl")
        
    
```

```python
mat = [[5,10,-5],[7,0,13],[10,8,-1]]
mat_list = i_list(mat)
poin = poincare_index(mat_list)
type_finder(poin)
print(poin)
```

```python
def ridge_pattern(A,f0,th):
    array = np.zeros((600, 600))
    for i in range(600):
        for j in range(600):
            #A = 80
            #f0 = 0.01
            theta = math.radians(th)
            inner_value = 2*np.pi*f0*(i*np.cos(theta)+j*np.sin(theta))
            w = A*np.cos(inner_value)
            array[i][j] = w

    plt.imshow(array, "gray")
    plt.show()
```

```python
A = 80
fo = 0.01
angles = [0,45,90,135]
for theta in angles:
    ridge_pattern(A,fo,theta)
```
